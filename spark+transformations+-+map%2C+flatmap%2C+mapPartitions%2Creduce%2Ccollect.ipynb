```python
one_through_9 = range(1,10)
parallel = sc.parallelize(one_through_9,3)#splitting range into 3 partiions
#define a function that needs to be applied on each partition
def sumPart(iterator):yield sum(iterator)
#pass function to execute on diff nodes
parallel.mapPartitions(sumPart).collect()
#partition 1: 1+2+3 = 6
#partition 2: 4+5+6 = 15
#partition 3: 7+8+9 = 24
```

```python
filtered_rows = baby_names.filter(lambda line: "Count" not in line).map(lambda line:line.split(","))
```

```python
#count per key
filtered_rows.map(lambda x : (str(x[1]),int(x[4]))).reduceByKey(lambda v1,v2:v1 + v2).collect()
```

```python
#sortByKey -asc
filtered_rows.map(lambda x:(str(x[1]),int(x[4]))).sortByKey().collect()
```

```python
#sortByKey - desc
filtered_rows.map(lambda x:(str(x[1]),int(x[4]))).sortByKey(False).collect()
```

```python
#reducebyKey + sortByKey -asc
filtered_rows.map(lambda x:(str(x[1]),int(x[4]))).reduceByKey(lambda v1,v2: v1 + v2).sortByKey().filter(lambda x : "GAVIN" in x).collect()
```
